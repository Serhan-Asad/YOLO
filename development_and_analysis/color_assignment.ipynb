```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
image_path = "/Users/qandeelasad/Desktop/Serhan/YOLO/output_videos/cropped_img.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
 
```

```python
plt.imshow(image)
plt.show
```

```python
top_half_imagae = image[0:int(image.shape[0]/2),:]
```

```python
image_2d = top_half_imagae.reshape(-1,3)
#kcluster with 2
kMeans = KMeans(n_clusters=2, random_state=0).fit(image_2d)
#get cluster label
labels = kMeans.labels_
#reshape to originial
clustered_image = labels.reshape(top_half_imagae.shape[0], top_half_imagae.shape[1])
#display clustered image
plt.imshow(clustered_image)
plt.show
```

```python
corner_clusters = [clustered_image[0,0], clustered_image[0,-1], clustered_image[-1,0], clustered_image[-1,-1]]
non_player_cluster = max(set(corner_clusters), key=corner_clusters.count)
```

```python
player_cluster = 1 - non_player_cluster 
print(player_cluster)
```

```python
kMeans.cluster_centers_[player_cluster]
```
